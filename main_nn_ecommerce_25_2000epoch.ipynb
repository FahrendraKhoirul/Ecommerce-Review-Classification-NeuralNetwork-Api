## Load Dataset from Pickle File


```python
import pickle
import numpy as np

with open('./Model/Dataset Skripsi - Fix Dataset (Final) - Upsampled - Clean - 90 - Train - TFIDF.pkl', 'rb') as f:
    df_train_90 = pickle.load(f)

with open('./Model/Dataset Skripsi - Fix Dataset (Final) - Upsampled - Clean - 90 - Test - TFIDF.pkl', 'rb') as f:
    df_test_90 = pickle.load(f)

with open('./Model/Dataset Skripsi - Fix Dataset (Final) - Upsampled - Clean - 80 - Train - TFIDF.pkl', 'rb') as f:
    df_train_80 = pickle.load(f)

with open('./Model/Dataset Skripsi - Fix Dataset (Final) - Upsampled - Clean - 80 - Test - TFIDF.pkl', 'rb') as f:
    df_test_80 = pickle.load(f)

with open('./Model/Dataset Skripsi - Fix Dataset (Final) - Upsampled - Clean - 70 - Train - TFIDF.pkl', 'rb') as f:
    df_train_70 = pickle.load(f)

with open('./Model/Dataset Skripsi - Fix Dataset (Final) - Upsampled - Clean - 70 - Test - TFIDF.pkl', 'rb') as f:
    df_test_70 = pickle.load(f)
```

```python
# 90 : 10
X_train_90 = df_train_90['tfidf']
y_train_90 = df_train_90['category_id']

X_test_90 = df_test_90['tfidf']
y_test_90 = df_test_90['category_id']

# 80 : 20
X_train_80 = df_train_80['tfidf']
y_train_80 = df_train_80['category_id']

X_test_80 = df_test_80['tfidf']
y_test_80 = df_test_80['category_id']

# 70 : 30
X_train_70 = df_train_70['tfidf']
y_train_70 = df_train_70['category_id']

X_test_70 = df_test_70['tfidf']
y_test_70 = df_test_70['category_id']
```

```python
# Make it suitable for my Neural Network input
# 90 : 10
X_train_90_np = np.array(X_train_90)
X_train_90_np = np.array([np.array(x) for x in X_train_90])
X_train_90_np_T = X_train_90_np.T

X_test_90_np = np.array(X_test_90)
X_test_90_np = np.array([np.array(x) for x in X_test_90])
X_test_90_np_T = X_test_90_np.T

# 80 : 20
X_train_80_np = np.array(X_train_80)
X_train_80_np = np.array([np.array(x) for x in X_train_80])
X_train_80_np_T = X_train_80_np.T

X_test_80_np = np.array(X_test_80)
X_test_80_np = np.array([np.array(x) for x in X_test_80])
X_test_80_np_T = X_test_80_np.T

# 70 : 30
X_train_70_np = np.array(X_train_70)
X_train_70_np = np.array([np.array(x) for x in X_train_70])
X_train_70_np_T = X_train_70_np.T

X_test_70_np = np.array(X_test_70)
X_test_70_np = np.array([np.array(x) for x in X_test_70])
X_test_70_np_T = X_test_70_np.T

```

```python
print('X_train_90_np_T.shape :', X_train_90_np_T.shape)
print('X_test_90_np_T.shape :', X_test_90_np_T.shape)
print('X_train_80_np_T.shape :', X_train_80_np_T.shape)
print('X_test_80_np_T.shape :', X_test_80_np_T.shape)
print('X_train_70_np_T.shape :', X_train_70_np_T.shape)
print('X_test_70_np_T.shape :', X_test_70_np_T.shape)
```

# Results


```python
import matplotlib.pyplot as plt

# plot the cost in one graph
plt.plot(cost_90, label='90 : 10')
plt.plot(cost_80, label='80 : 20')
plt.plot(cost_70, label='70 : 30')
plt.xlabel('epoch')
plt.ylabel('cost')
plt.title('Cost of Training')
plt.legend()
```

```python
# plot the accuracy in one graph
plt.plot(acc_90, label='90 : 10')
plt.plot(acc_80, label='80 : 20')
plt.plot(acc_70, label='70 : 30')
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.title('Accuracy of Training')
plt.legend()
```

```python
# make confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns

# 90 : 10
conf_matrix_90 = confusion_matrix(y_test_90, predictions_90)

# 80 : 20
conf_matrix_80 = confusion_matrix(y_test_80, predictions_80)

# 70 : 30
conf_matrix_70 = confusion_matrix(y_test_70, predictions_70)

```

```python
# plot the confusion matrix
import matplotlib.colors as mcolors

cat_code = {0: 'Produk', 1: 'Layanan Pelanggan', 2: 'Pengiriman'}
# 90 : 10
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_90, annot=True, fmt='d', xticklabels=cat_code.values(), yticklabels=cat_code.values(), cmap="Blues",
            annot_kws={"size": 16})
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.title('Confusion Matrix of 90 : 10')

# 80 : 20
plt.figure(figsize=(10, 7))
sns.heatmap(conf_matrix_80, annot=True, fmt='d', xticklabels=cat_code.values(), yticklabels=cat_code.values(), cmap="Blues",
            annot_kws={"size": 16})
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.title('Confusion Matrix of 80 : 20')

# 70 : 30
plt.figure(figsize=(10, 7))
# make a bigger text in the heatmap
sns.heatmap(conf_matrix_70, annot=True, fmt='d', xticklabels=cat_code.values(), yticklabels=cat_code.values(), cmap="Blues",
            annot_kws={"size": 16})
plt.xlabel('Prediction')
plt.ylabel('Actual')
plt.title('Confusion Matrix of 70 : 30')

```

```python
# calculate the precision, recall, and F1
from sklearn.metrics import precision_score, recall_score, f1_score

# 90 : 10
precision_90 = precision_score(y_test_90, predictions_90, average=None)
recall_90 = recall_score(y_test_90, predictions_90, average=None)
f1_90 = f1_score(y_test_90, predictions_90, average=None)

# 80 : 20
precision_80 = precision_score(y_test_80, predictions_80, average=None)
recall_80 = recall_score(y_test_80, predictions_80, average=None)
f1_80 = f1_score(y_test_80, predictions_80, average=None)

# 70 : 30
precision_70 = precision_score(y_test_70, predictions_70, average=None)
recall_70 = recall_score(y_test_70, predictions_70, average=None)
f1_70 = f1_score(y_test_70, predictions_70, average=None)

# print the result
# 90 : 10
print('\n90 : 10')
print('Accuracy  :', accuracy_90)
print('Precision :', precision_90)
print('Recall    :', recall_90)
print('F1        :', f1_90)

# 80 : 20
print('\n80 : 20')
print('Accuracy  :', accuracy_80)
print('Precision :', precision_80)
print('Recall    :', recall_80)
print('F1        :', f1_80)

# 70 : 30
print('\n70 : 30')
print('Accuracy  :', accuracy_70)
print('Precision :', precision_70)
print('Recall    :', recall_70)
print('F1        :', f1_70)


```

```python
# plot each precision, recall, and F1 in one graph

# 90 : 10
precision_90_avg = np.mean(precision_90)
recall_90_avg = np.mean(recall_90)
f1_90_avg = np.mean(f1_90)


# 80 : 20
precision_80_avg = np.mean(precision_80)
recall_80_avg = np.mean(recall_80)
f1_80_avg = np.mean(f1_80)

# 70 : 30
precision_70_avg = np.mean(precision_70)
recall_70_avg = np.mean(recall_70)
f1_70_avg = np.mean(f1_70)

fig, ax = plt.subplots()
index = np.arange(3)
bar_width = 0.3

rects1 = plt.bar(index, [precision_90_avg, recall_90_avg, f1_90_avg], bar_width,  color='C0', label='90 : 10')
rects2 = plt.bar(index + bar_width, [precision_80_avg, recall_80_avg, f1_80_avg], bar_width,  color='C1', label='80 : 20')
rects3 = plt.bar(index + bar_width*2, [precision_70_avg, recall_70_avg, f1_70_avg], bar_width,  color='C2', label='70 : 30')

# Add the exact numbers on top of each bar
for i, v in enumerate([precision_90_avg, recall_90_avg, f1_90_avg]):
    plt.text(i - 0.15, v + 0.01, str(round(v, 3)))
for i, v in enumerate([precision_80_avg, recall_80_avg, f1_80_avg]):
    plt.text(i + bar_width - 0.15, v + 0.01, str(round(v, 3)))
for i, v in enumerate([precision_70_avg, recall_70_avg, f1_70_avg]):
    plt.text(i + bar_width*2 - 0.15, v + 0.01, str(round(v, 3)))

plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.title('Scores by Category')
plt.xticks(index + bar_width, ('Precision', 'Recall', 'F1'))
plt.legend(loc='lower right')

plt.show()
print('90 : 10')
print('Precision  :', precision_90_avg)
print('Recall     :', recall_90_avg)
print('F1         :', f1_90_avg)

print('80 : 20')
print('Precision  :', precision_80_avg)
print('Recall     :', recall_80_avg)
print('F1         :', f1_80_avg)

print('70 : 30')
print('Precision  :', precision_70_avg)
print('Recall     :', recall_70_avg)
print('F1         :', f1_70_avg)
```
