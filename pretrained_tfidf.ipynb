```python
import pickle
import pandas as pd
from tfidf import TFIDF
```

```python
df = pd.read_csv('../../Dataset/Dataset Skripsi - Fix Dataset (Final).csv')
df
```

```python
df_sample = df.sample(frac=0.8, random_state=1)
df_sample
```

```python
tfidf = TFIDF(df_sample.text)
```

```python
# save the model to disk
filename = 'tfidf_model.pkl'
pickle.dump(tfidf, open(filename, 'wb'))
```
