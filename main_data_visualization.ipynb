# Load All Data


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Load add Cost and Accuracy data

cost_acc_A1 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_A1.csv')
cost_acc_A2 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_A2.csv')
cost_acc_A3 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_A3.csv')
cost_acc_B1 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_B1.csv')
cost_acc_B2 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_B2.csv')
cost_acc_B3 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_B3.csv')
cost_acc_C1 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_C1.csv')
cost_acc_C2 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_C2.csv')
cost_acc_C3 = pd.read_csv('../../Dataset/Cost_Accuracy_Model_C3.csv')

# load result summary
result_summary = pd.read_excel('../../Dataset/Result Summary.xlsx', sheet_name='clean')
```

```python
def plot_all_model(models, cost_or_acc, title):
    metric = 'accuracy' if cost_or_acc == 'accuracy' else 'cost'
    plt.figure(figsize=(10, 6))
    labels = ['A1', 'A2', 'A3', 'B1', 'B2', 'B3', 'C1', 'C2', 'C3']
    color_for_A = 'red'
    color_for_B = 'green'
    color_for_C = 'blue'
    # for index, model in enumerate(models):
    #     if index < 3:
    #         plt.plot(model[metric], label=labels[index], color=color_for_A, linewidth=1)
    #     elif index < 6:
    #         plt.plot(model[metric], label=labels[index], color=color_for_B, linewidth=1)
    #     else:
    #         plt.plot(model[metric], label=labels[index], color=color_for_C, linewidth=1)
    for index, model in enumerate(models):
        plt.plot(model[metric], label=labels[index], linewidth=1)
        plt.text(len(model[metric])-1, model[metric].iloc[-1], labels[index])
    plt.xlabel('Epoch')
    plt.ylabel(metric)
    plt.title(title)
    plt.legend()
    plt.show()
```

```python
models = [cost_acc_A1, cost_acc_A2, cost_acc_A3, cost_acc_B1, cost_acc_B2, cost_acc_B3, cost_acc_C1, cost_acc_C2, cost_acc_C3]
plot_all_model(models, 'accuracy', 'Accuracy of All Models')
```

```python
#  plot all last epoch accuracy
last_epoch_accuracy = []

for model in models:
    last_epoch_accuracy.append(model['accuracy'].iloc[-1])

fig, ax = plt.subplots()
index = np.arange(3)
bar_width = 0.3

plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Last Epoch Accuracy of All Models')
plt.show()

```
